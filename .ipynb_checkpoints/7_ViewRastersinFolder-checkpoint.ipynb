```python
import bokeh
import random

def rand_cmap():

    c_list = list(sorted(bokeh.palettes.all_palettes.keys()))
    stop = len(c_list)
    i = random.randrange(stop)
    return(c_list[i])
```

```python
import xarray as xr

def img_plot_proxy(img):
    
    #cmap=rand_cmap()
    #print(cmap)
    cmap='RdYlGn'
    da = xr.open_rasterio(img)
    da_masked = da.where(da<10000.0)
    da_masked = da_masked.where(da>0.0)
    img_figure = da_masked.hvplot.image(dynamic=False,rasterize=True,grid=True, clim=(0,1),
                                        invert=False, width=400, height=500, cmap=cmap)
    #img_figure = da_masked.plot()
    print("why did you call me")
    return(img_figure)
```

```python
title = '## Tif Display Panel App'

tif_select = pn.widgets.Select(name='tif1', options=tif_list)

@pn.depends(tif_select.param.value)
def refresh_holoview(tif1):
    return(img_plot_proxy(tif_select.value))
```

```python
title2='### Tif View'
row = pn.Row(
    pn.Column(title2,  pn.panel(refresh_holoview,width=500, height=400), tif_select)
)

row.servable()
```
